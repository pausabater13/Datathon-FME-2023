#Load environment

```
environment_name = 'CartPole-v1'
env = gym.make(environment_name, render_mode="human")
```

```
episodes = 5 #an episode is a full game within the environment
for episode in range(1, episodes+1):
    state = env.reset() #Get initial set of observations
    terminated = False
    score = 0 
    
    while not terminated:
        env.render() #visualize environment graph representation. Donesn´t work in collabs
        #OBSERVATION SPACE: 
            #env.observation_space returns (lower_bound, upper_bound, nº elems, type)
            #env.observation_space.sample() returns (cart_position, cart_velocity, pole_angle, pole_velocity)
        
        #ACTION SPACE
            #env.action_space = Discrete(2)
        action = env.action_space.sample()#Returns a sample from the action_space

        n_state, reward, terminated, truncated, info = env.step(action) #apply an action to the environment
        score+=reward
    print('Episode:{} Score:{}'.format(episode, score))
env.close() #Close down the render frame

#Understanding environment https://github.com/openai/gym/blob/master/gym/envs/classic_control/cartpole.py
```

```
import os
save_path = os.path.join('Training', 'Saved Models')
log_path = os.path.join('Training', 'Logs')
```

```
env = gym.make(environment_name)
env = DummyVecEnv([lambda: env])
```

#Adding a callback to the training stage

```
from stable_baselines3.common.callbacks import EvalCallback, StopTrainingOnRewardThreshold
```

```
stop_callback = StopTrainingOnRewardThreshold(reward_threshold=190, verbose=1)
eval_callback = EvalCallback(env, 
                             callback_on_new_best=stop_callback, 
                             eval_freq=10000, #Evaluation each 10000 training cases
                             best_model_save_path=save_path, 
                             verbose=1)
```

```
model = PPO('MlpPolicy', env, verbose = 1, tensorboard_log=log_path)#same as in training model step
```

```
model.learn(total_timesteps=20000, callback=eval_callback)
```

```
model_path = os.path.join('Training', 'Saved Models', 'best_model')
model = PPO.load(model_path, env=env)
```

```
evaluate_policy(model, env, n_eval_episodes=10, render=True)
```

```
env.close()
```
